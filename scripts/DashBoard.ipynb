```python
import os
import sklearn
import sklearn.datasets
import sklearn.ensemble
import xgboost
import numpy as np
import pandas as pd
import pickle

import sys  
sys.path.insert(0, '../src/')
import featureExtraction as fe

from omnixai.data.tabular import Tabular
from omnixai.preprocessing.tabular import TabularTransform
from omnixai.explainers.tabular import TabularExplainer
from omnixai.visualization.dashboard import Dashboard

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('../data/current-dataset-unnormalized/feature-dataset.csv', delimiter=",",index_col=0)
# Load unnormalized dataset
with open('../data/'+'fe_dataset.pkl', 'rb') as file:
    fe_dataset = pickle.load(file)

tabular_data = Tabular(
    data=df,
    target_column='label'
)
print(tabular_data)
```

```python
transformer = TabularTransform().fit(tabular_data)
class_names = transformer.class_names
x = transformer.transform(tabular_data)
```

```python
from sklearn.ensemble import RandomForestClassifier

transformer = TabularTransform().fit(tabular_data)
class_names = transformer.class_names
x = transformer.transform(tabular_data)

#CUSTOM SPLIT
train =  x[fe_dataset.idx_train,:-1]
test =  x[fe_dataset.idx_test,:-1]
train_labels =  x[fe_dataset.idx_train,-1]
test_labels =  x[fe_dataset.idx_test,-1]

#STANDARD SPLIT
# train, test, train_labels, test_labels = \
#     sklearn.model_selection.train_test_split(x[:, :-1], x[:, -1], train_size=0.80)

print('Training data shape: {}'.format(train.shape))
print('Test data shape:     {}'.format(test.shape))

# gbtree = RandomForestClassifier()
gbtree = xgboost.XGBClassifier(n_estimators=300, max_depth=50)
gbtree.fit(train, train_labels)
print('Test accuracy: {}'.format(
    sklearn.metrics.accuracy_score(test_labels, gbtree.predict(test))))

# Convert the transformed data back to Tabular instances
train_data = transformer.invert(train)
test_data = transformer.invert(test)
```

```python
print('Test accuracy: {}'.format(
    sklearn.metrics.accuracy_score(test_labels, gbtree.predict(transformer.transform(test_data)))))
```

```python
preprocess = lambda z: transformer.transform(z)
```

```python
# Initialize a TabularExplainer
explainers = TabularExplainer(
    explainers=["lime","shap","ale"],
    mode="classification",
    data=train_data,
    model=gbtree,
    preprocess=preprocess,
    params={
        "lime": {"kernel_width": 3},
        "shap": {"nsamples": 100}
        
    }
)
```

```python
car_sample = 16196
walk_sample = 6201
confusion_car_sample = 1181
```

```python
# Generate explanations
test_instances = test_data[confusion_car_sample]
local_explanations = explainers.explain(X=test_instances)
global_explanations = explainers.explain_global(
    # params={"ale": {"features": ["gyro_x_zero-crossing","acc_y_mean-mags","acc_z_zero-crossing","mag_x_zero-crossing","mag_y_zero-crossing","mag_z_zero-crossing"]}}
    params={"ale": {"features": ["gyro_x_zero-crossing","mag_z_zero-crossing"]}}

)
```

```python
index=0
print("LIME results:")
local_explanations["lime"].ipython_plot(index, class_names=class_names)
print("SHAP results:")
local_explanations["shap"].ipython_plot(index, class_names=class_names)
print("ALE results:")
global_explanations["ale"].ipython_plot(class_names=class_names)
```

```python
# Compute metrics
from omnixai.explainers.prediction import PredictionAnalyzer

analyzer = PredictionAnalyzer(
    mode="classification",
    test_data=test_data,
    test_targets=test_labels,
    model=gbtree,
    preprocess=preprocess
)
prediction_explanations = analyzer.explain()
```

```python
# Launch a dashboard for visualization
dashboard = Dashboard(
    instances=test_instances,
    local_explanations=local_explanations,
    global_explanations=global_explanations,
    prediction_explanations=prediction_explanations,
    class_names=class_names,
    explainer=explainers
)
dashboard.show()
```
